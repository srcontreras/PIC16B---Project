```python
import pandas as pd
import numpy as np
from yahoo_fin import stock_info as si
from matplotlib import pyplot as plt
from plotly import express as px

company = si.get_data('sbux')
company = company.loc['2021-01-04':'2023-11-14'] 
```

```python
company
```

```python
company["date"] = company.index
```

```python
company = company.reset_index()
```

```python
company["yesterday"] = company["close"].shift(1) 
company = company.fillna(company["close"][0])
company["%change"] = (company["close"]- company["yesterday"]) / company["yesterday"]
company
```
